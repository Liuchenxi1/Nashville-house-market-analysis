```python
def fix_columns(column_list):
    new_cols =[]
    for col in column_list:
        col = col.lower()
        col = col.replace(', tn','')
        col = col.replace(', ca','')
        col = col.replace(', ga','')
        new_cols.append(col)
    
    return new_cols 
## Reshapre the Columns names by dorpping ", TN"
```
