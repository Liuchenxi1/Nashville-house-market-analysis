```python
threecities = rental[(rental['RegionName'] == 'Los Angeles-Long Beach-Anaheim, CA') | (rental['RegionName'] == 'Atlanta, GA') | (rental['RegionName'] == 'Nashville, TN')]
```

```python
threecities.head()
```

```python
cities_new = threecities.transpose()
cities_new
```

```python
cities_new.columns = cities_new.iloc[0]
cities_new
```

```python
cities_new = cities_new.iloc[1: , :]
cities_new
```

```python
cities_new.columns
```

```python
cities_new = cities_new.rename(columns ={'index':'yearanddate'})
```

```python
cities_new.columns
```

```python
def fix_columns(column_list):
    new_cols =[]
    for col in column_list:
        col = col.lower()
        col = col.replace(', tn','')
        col = col.replace(', ca','')
        col = col.replace(', ga','')
        new_cols.append(col)
    
    return new_cols 
```

```python
cities_new.columns = fix_columns(cities_new.columns)
```

```python
cities_new = cities_new.rename(columns = {'los angeles-long beach-anaheim':'la'})
```

```python
cities_new.columns
```

```python
cities_new = cities_new.reset_index()
cities_new
```

```python
cities_new.columns
```

```python
cities_new[['year','date']] = cities_new['index'].str.split("-", n=1, expand = True)
```

```python
cities_new.head()
```

```python
cities_new = cities_new.drop(['index','date'], axis =1)
```

```python
cities_new[['la','atlanta','nashville','year']] = cities_new[['la','atlanta','nashville','year']].apply(pd.to_numeric)
```

```python
rentalbyyear = cities_new.groupby(['year'])['la','atlanta','nashville'].median().reset_index()
print(rentalbyyear)
```

```python
rentalbyyear = rentalbyyear.reset_index()
```

```python
rentalbyyear.plot(x="year", y=["la", "atlanta", "nashville"], kind="bar", color = ('green', 'darkorange','yellow'))
```

```python
rentalbyyear
```

```python
rentalbyyear = rentalbyyear.drop(['index'], axis =1)
```

```python
plt.plot('year','la',data=rentalbyyear, color='skyblue',linewidth = 2)
plt.plot('year', 'atlanta', data=rentalbyyear, color= 'black',linewidth = 2)
plt.plot('year', 'nashville', data=rentalbyyear, color='yellow',linewidth = 2)
##plt.locator_params(axis='x', nbins=10)

plt.xlabel('year')

plt.legend()

plt.show()
```
